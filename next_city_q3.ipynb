In this case, it looks the size of the data is too small. Therefore, including features, such as joining date and userid will lead to overfitting and we may end up to a bad model, with low performance.

```python
from __future__ import print_function

import pandas as pd
import numpy as np 
import sys
import io

# import json to  read json file
import json
from pandas.io.json import json_normalize

import unidecode

#import importlib
import next_city
import missing_country
# importlib.reload(next_city)
# importlib.reload(missing_country)
from next_city import * 
from missing_country import *

pd.options.mode.chained_assignment = None
```

```python
json_file = "city_search.json"
world_cities_csv = "world-cities1.csv"
user_ucountry = True

# Create dataframe from the json file
dframe = load_data_as_df(json_file)
preprocess_data(dframe)
```

```python
missing_c = missining_country_likely(json_file, world_cities_csv)
missing_country = missing_c["cn"]
missing_country
```

```python
replace_missing_country_for_users(dframe["user"], missing_country)
```

```python
# Get users' contries. 
countriesSet = find_user_countries_set(dframe)
print(countriesSet)
```

```python
# lets split the cities, to make it as a list of cities. Not replacing to cities column, just to keep the original column. otherwise, it would be just just replaced. 
dframe["cs"] = dframe["cities"].apply(lambda x: x[0].split(', '))
```

```python
# finding user country , to train the model with that included...
dframe["user-country"] = dframe["user"].apply(lambda x: [x["country"]])
```

```python
# adding user country to the beginning of the list
dframe["cs"] = dframe["user-country"] + dframe["cs"]
```

```python
# Not required anymore
citiesSearchedSeries = dframe["cs"]
```

```python
# Creating index for cities and countries
citiesIndex = list(set(item for sublist in citiesSearchedSeries for item in sublist[1:]))
citiesIndex = citiesIndex + list(countriesSet) + list({''})
```

```python
Tx = 14
n_x = len(citiesIndex)
print("Tx: {} & n_x: {}".format(Tx, n_x))
```

```python
# Building Data ...
print("Creating training set...")
X, X_test, Y, Y_test = build_data(citiesSearchedSeries, Tx, user_ucountry)

print(type(X))

# Vectorizing ...
print("Vectorizing training set...")
x, y = vectorization(X, Y, len(citiesIndex), citiesIndex, Tx)

# Loading Model ...
print("Loading model...")
model = city_model((Tx, n_x,), len(citiesIndex))
model.summary()
```

```python
# Compiling Model 
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# training Model 
print_callback = LambdaCallback(on_epoch_end=on_epoch_end)
model.fit(x, y, batch_size=128, epochs=5, callbacks=[print_callback])
```

```python
# Validating ...
preds = model.predict(x[0:1], verbose=0)[0]
next_index = sample(preds, citiesIndex, temperature = 1.0)
next_city = citiesIndex[next_index]
```
